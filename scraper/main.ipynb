# Scraping Bike Information - Decathlon

```python
# Bikes classified as women's bikes

from selenium import webdriver
from selenium.webdriver.common.by import By

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# URL to scrape
base_url = "https://www.decathlon.fr/tous-les-sports/velo-cyclisme/velos-femme"

# List to store all product information
bike_information_women = []

# Loop through all pages
for page_number in range(1, 25):  # 24 pages in total
    # Construct page URL
    page_url = f"{base_url}?page={page_number}"
    
    # Navigate to the page
    driver.get(page_url)
    
    # Wait for the page to load
    driver.implicitly_wait(10)  # Adjust the wait time as needed
    
    # Find all elements with the specified class for product links
    product_links = driver.find_elements(By.CLASS_NAME, "dpb-product-model-link")
    
    # Find all elements with the specified class for product names
    product_names = driver.find_elements(By.CSS_SELECTOR, 'h2.vtmn-typo_caption-1')
    
    # Find all elements with the specified class and aria-label for product prices
    product_prices = driver.find_elements(By.CSS_SELECTOR, 'span.vtmn-price[aria-label="Prix"]')
    
    # Extract and store the information
    for link, name, price in zip(product_links, product_names, product_prices):
        product_link = link.get_attribute("href")
        product_name = name.get_attribute("innerText").strip()
        product_price = price.get_attribute("innerText").strip()
        
        # Store the information in a dictionary
        product_info = {
            "product_link": product_link,
            "product_name": product_name,
            "product_price": product_price
        }
        
        # Append the dictionary to the list
        bike_information_women.append(product_info)

# Quit the WebDriver
driver.quit()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Let's assume we have two dataframes: bike_information_men_df and bike_information_women_df

# Extract the product names from each dataframe
men_product_names = set(bike_information_men_df['product_name'])
women_product_names = set(bike_information_women_df['product_name'])

# Calculate the counts
total_unique_bikes = len(men_product_names.union(women_product_names))
only_men = len(men_product_names - women_product_names)
only_women = len(women_product_names - men_product_names)
both = len(men_product_names & women_product_names)

# Calculate percentages
percentage_men = (only_men / total_unique_bikes) * 100
percentage_women = (only_women / total_unique_bikes) * 100
percentage_both = (both / total_unique_bikes) * 100

# Create a new dataframe for visualization
bike_comparison_df = pd.DataFrame({
    'Category': ['Only Men', 'Only Women', 'Both'],
    'Percentage': [percentage_men, percentage_women, percentage_both]
})

# Define colors
colors = ['#f0ad2f', '#8724f5', 'grey']

# Plot the results
plt.figure(figsize=(8, 6))
barplot = sns.barplot(x='Category', y='Percentage', data=bike_comparison_df, palette=colors)
plt.title('Bike Comparison Between Men and Women')
plt.xlabel('Category')
plt.ylabel('Percentage of Total Unique Bikes')
plt.ylim(0, 100)

# Add data callouts
for index, row in bike_comparison_df.iterrows():
    barplot.text(index, row['Percentage'] + 1, f"{row['Percentage']:.2f}%", color='black', ha="center")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Function to clean the product names
def clean_product_name(name):
    # Remove special characters and convert to lowercase
    cleaned_name = re.sub(r'[^a-zA-Z0-9\s]', '', name).lower()
    return cleaned_name

# Assume we have two dataframes: bike_info_men_df and bike_info_women_df

# Let's extract the product names from each dataframe and clean them
men_product_names = set(bike_information_men_df['product_name'].apply(clean_product_name))
women_product_names = set(bike_information_women_df['product_name'].apply(clean_product_name))

# Calculate the counts
total_unique_bikes = len(set(men_product_names).union(set(women_product_names)))
only_men = len(set(men_product_names) - set(women_product_names))
only_women = len(set(women_product_names) - set(men_product_names))
both = len(set(men_product_names) & set(women_product_names))

# Calculate percentages
percentage_men = (only_men / total_unique_bikes) * 100
percentage_women = (only_women / total_unique_bikes) * 100
percentage_both = (both / total_unique_bikes) * 100

# Create a new dataframe for visualization
bike_comparison_df = pd.DataFrame({
    'Category': ['Only Men', 'Only Women', 'Both'],
    'Percentage': [percentage_men, percentage_women, percentage_both]
})

# Define colors
colors = ['#f0ad2f', '#8724f5', 'grey']

# Plot the results
plt.figure(figsize=(8, 6))
barplot = sns.barplot(x='Category', y='Percentage', data=bike_comparison_df, palette=colors)
plt.title('Bike Comparison Between Men and Women')
plt.xlabel('Category')
plt.ylabel('Percentage of Total Unique Bikes')
plt.ylim(0, 100)

# Add data callouts
for index, row in bike_comparison_df.iterrows():
    barplot.text(index, row['Percentage'] + 1, f"{row['Percentage']:.2f}%", color='black', ha="center")

plt.show()
```
